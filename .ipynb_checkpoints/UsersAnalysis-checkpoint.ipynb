# Users Analysis

## Why is this project valuable?

With enough time, energy, data, and resources, this database can provide crucial information to bookstores about their customers and their preferences. 

Import the libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import pandas as pd
import ast
```

Load the users dataset:

```python
users = pd.read_csv(r'users.csv')
user_df = pd.DataFrame(users)
book_df = pd.read_csv(r"Amazon_books_cleaned.csv")
```

A function to turn one users reading history into a dataframe:

```python
def book_hist_to_df(user_df, target_user_ID):
    """This function takes a row from the users database and returns the users reading history in a DataFrame."""    
    book_ratings = []
    for book in ast.literal_eval(user_df.iloc[target_user_ID]["book_history"]):
            book_ratings.append({"user_id": target_user_ID, "book": book["book"], "rating": book["rating"], 'genre': book['genre']})
    return pd.DataFrame(book_ratings)
```

For the sake of this analysis, let's choose one random reader. We can choose the one with ID: 5 and print their details:

```python
target_user_id = 5
user_books_df = book_hist_to_df(user_df, target_user_id)

print(f"User Name: {user_df.iloc[target_user_id]['name']}\nAge: {user_df.iloc[target_user_id]['age']}\nPrefererred Genres: {user_df.iloc[target_user_id]['preferences']}")
```

We can do a pie chart to analyze the readers most read genres:

```python
# Plotting the distribution of genres (Pie chart)
genre_counts = user_books_df['genre'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Genres in Userâ€™s Reading History')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

It is also possible to visualize a reader's change in ratings over time, depending on the book they read.

```python
plt.figure(figsize=(8, 4))
plt.plot(user_books_df['rating'], marker='o', linestyle='-', color='b')
plt.title('Change in User Ratings Over Time')
plt.xlabel('Book Index (Chronological Order)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

This is an example so that we can visualize how powerful of a tool this is in order to understand readers, what they enjoy and what they don't, and also what books can be recommended to them.
Knowing this, a book store can gain invaluable insight into their customers and understand the market.

We can also visualize information about all users.

```python
# Some debugging
def parse_book_history(book_history_str):
    try:
        return ast.literal_eval(book_history_str)  # Try to safely evaluate the string
    except (ValueError, SyntaxError):
        return []  # Return an empty list if parsing fails
user_df['book_history'] = user_df['book_history'].apply(parse_book_history)    
```

Pie chart of all availabe genres across all users:

```python
all_genres = []
for index, row in user_df.iterrows():
    for book in row['book_history']:
        all_genres.append(book['genre'])

# Plot the genre distribution (Pie chart)
genre_counts = pd.Series(all_genres).value_counts()

plt.figure(figsize=(8, 6))
plt.pie(genre_counts, labels=genre_counts.index, startangle=140)
plt.title('Distribution of Genres Across All Users')
plt.axis('equal')  
plt.show()
```

We can also plot the distribution of ratings done by the users:

```python
all_ratings = []
for index, row in user_df.iterrows():
    for book in row['book_history']:
        all_ratings.append(book['rating'])

# Plot histogram of ratings
plt.figure(figsize=(5, 4))
plt.hist(all_ratings, bins=10, color='skyblue', edgecolor = 'white')
plt.title('Distribution of Ratings Across All Users')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

We can plot the most rated books, aka, the books most read:

```python
# Count how many times each book was rated
book_counts = pd.Series([book['book'] for row in user_df['book_history'] for book in row]).value_counts()

# Plot most rated books
book_counts.head(10).plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Most Rated Books')
plt.xlabel('Number of Ratings')
plt.ylabel('Book')
plt.show()
```
